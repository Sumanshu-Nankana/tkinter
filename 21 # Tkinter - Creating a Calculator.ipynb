```python
from tkinter import *

root = Tk()

root.geometry("370x500")
root.title("Calculator")

####   Output Screen on TOP ####
out_value = StringVar()
out_value.set("")
out_screen = Entry(root,textvariable=out_value,font="lucida 30 bold")
out_screen.pack(fill=X,padx=10,pady=10,ipadx=10) #ipadx - is for internal padding

all_buttons = ['9','8','7','+','6','5','4','*','3','2','1','/','C','0','=','-','%','.','00','^']

### Functions ###
def click(event):
    
    global out_value
    text = event.widget.cget("text")
    
    if text == '=': 
        if out_value.get().isdigit():
            value = int(out_value.get())
        else:
            try:
                value = eval(out_value.get())
            except Exception as e:
                print(e)
                value="Error"
        out_value.set(value)
        out_screen.update()
    elif text == 'C':
        out_value.set("")
        out_screen.update()
    else:
        out_value.set(out_value.get()+text)
        out_screen.update() 

#### We will create FRAMES and we will pack 4 buttons in 1 Frame ####
k = 0
for i in range(5):
    f1 = Frame(root,bg='grey')
    for j in range(4):
        b = Button(f1,text=all_buttons[k],font="lucida 20 bold",border=0,relief=SUNKEN)
        #b.pack(side=LEFT,padx=10,pady=10)
        b.pack(side=LEFT,expand=True,fill='both')
        b.bind('<Button-1>',click)
        k+=1
    f1.pack(expand=True,fill='both')

root.mainloop()
```

<h4> METHOD-2 -Some issue while converting integer to string </h4>

```python
from tkinter import *
import tkinter.messagebox as tmsg

class calculator(Tk):
    def __init__(self):
        super().__init__()
        self.geometry("400x400+0+0")  # 0+0 - means my GUI will open at top corner in my laptop.
        self.resizable(0,0)          #  We can't maximize and minimize it.
        self.title("Calculator")
        self.val = ""
        self.data = StringVar()
        self.a = 0
        self.b = 0
        self.oper = ""
        self.val2 = ""
    
    def clicked(self,event):
        text =  event.widget.cget("text")
        if text == 'C':
            self.val = ""
            self.a = 0
            self.b = 0
            self.oper = ''
            self.data.set(self.val)
        elif text == '+':
            self.a = int(self.val)
            self.oper = '+'
            self.val = self.val + self.oper
            self.data.set(self.val)
        elif text == '-':
            self.a = int(self.val)
            self.oper = '-'
            self.val = self.val + self.oper
            self.data.set(self.val)
        elif text == '*':
            self.a = int(self.val)
            self.oper = '*'
            self.val = self.val + self.oper
            self.data.set(self.val)
        elif text == '/':
            self.a = int(self.val)
            self.oper = '/'
            self.val = self.val + self.oper
            self.data.set(self.val)
        elif text == '=':
            self.val2 = self.val
            if self.oper == '+':
                self.b = int(self.val2.split('+')[1])
                c = self.a + self.b
                self.data.set(c)
                self.val = str(c)
            if self.oper == '-':
                self.b = int(self.val2.split('-')[1])
                self.c = self.a - self.b
                self.data.set(self.c)
                self.val = str(self.c)
            if self.oper == '*':
                self.b = int(self.val2.split('*')[1])
                self.c = self.a * self.b
                self.data.set(self.c)
                self.val = str(self.c)
            if self.oper == '/':
                self.b = int(self.val2.split('/')[1])
                if self.b == 0:
                    tmsg.showerror("Error","Division By Zero Not Allowed!!")
                    self.a = 0
                    self.b = 0
                    self.c = 0
                    self.val = ""
                    self.data.set(self.val)
                else:
                    self.c = int(self.a/self.b)
                    self.data.set(self.c)
                    self.val = str(self.c)
        else:
            self.val = self.val + text
            self.data.set(self.val)
    
    def create_design(self):
        self.data = StringVar()
        lb = Label(self,text=self.val,font="Verdana 22",anchor=SE,
                   textvariable=self.data,bg='#ffffff',fg='#000000').pack(expand=True,fill='both')
        all_keys = ['7','8','9','*','4','5','6','-','1','2','3','+','C','0','=','/']
        
        k = 0
        
        for i in range(4):
            f1 = Frame(self)
            for j in range(4):
                b = Button(f1,text=all_keys[k],font = "Verdana 20",relief=GROOVE,border=0)
                b.pack(side=LEFT,expand=True,fill="both")
                b.bind("<Button-1>",self.clicked )
                k+=1
            f1.pack(expand=True,fill="both")
        
if __name__ == "__main__":
    calc = calculator()
    calc.create_design()
    calc.mainloop()
```
