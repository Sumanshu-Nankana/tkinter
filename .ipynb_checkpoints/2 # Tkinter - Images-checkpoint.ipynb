```python
#To create any GUI in Python, There are so many modules

# Tkinter
# PyQt (Qt owned by Nokia)
# Kivy

# But we will use tkinter, it comes along with Python, There is no need to install seprately

# Let's import everything from tkinter - though it's not good thing
# but here we are learning basics so let's import everything
from tkinter import *

# create a root window using Tk() class
root = Tk()

# Now run a mainloop() function which is necssary - this provide a basic window for GUI
# In which we place our widgets
# But if we directly run root.mainloop() - a window open with some default size

# But we want, when first time window open, it open in some specific size
# So we need to set geometry parameter and in this we give width x height

# width x height
root.geometry("500x500")

# But Still we can manually drag the window size and increase it's width and height
# Let's fix it's min and max size.
root.minsize(200,100)
root.maxsize(300,300)

#Lets create a text label
label1 = Label(text="Label1")

#But we need to pack the variable to show it on GUI, If we not ack it, it will be not visible.
label1.pack()

#Let's display an image into GUI
photo_png = PhotoImage(file='1.png')

# Now put this into a label
photo_png_label = Label(image=photo_png)

# Now to see that in GUI, first pack the label
photo_png_label.pack()


# #Many image format not supported by Tkinter,example - jpg file format
# #so we can install pillow module 
# # sudo apt-get install python3-pil python3-pil.imagetk
# # pillow (Python Imaging Library - PIL) and import it

from PIL import Image,ImageTk
image = Image.open('2.jpg')
photo_jpg = ImageTk.PhotoImage(image)
photo_jpg_label = Label(image=photo_jpg)
photo_jpg_label.pack()

root.mainloop()
```

```python
#Many image format not supported by Tkinter,example - jpg file format
#so we can install pillow module 
# sudo apt-get install python3-pil python3-pil.imagetk
# pillow (Python Imaging Library - PIL) and import it

from tkinter import *
from PIL import Image,ImageTk

root = Tk()
root.geometry("500x500")

root.minsize(200,100)
root.maxsize(300,300)

image = Image.open('2.jpg')
photo_jpg = ImageTk.PhotoImage(image)
photo_jpg_label = Label(image=photo_jpg)
photo_jpg_label.pack()

root.mainloop()
```

```python
# Read the images from a folder using os module and show in a GUI
import os
from tkinter import *
from PIL import Image, ImageTk

path = "Images/"

all_images = os.listdir(path)

root = Tk()
root.geometry("800x800")
root.minsize(800,800)
root.maxsize(800,800)

for image in all_images:
    image_path = path + image
    image = Image.open(image_path)
    photo = ImageTk.PhotoImage(image)
    photo_label = Label(image=photo)
    photo_label.pack()

root.mainloop()
```
